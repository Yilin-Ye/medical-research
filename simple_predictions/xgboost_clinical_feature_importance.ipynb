```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from xgboost import plot_importance
import matplotlib.pyplot as plt
from numpy import sort
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import LeaveOneOut
```

```python
import os
os.chdir('C://Users//YE/Desktop/')
```

```python
combined_clinical=pd.read_csv("combined_clinical.csv")
```

# Patient

```python
plt.rcParams["font.family"] = "Arial"
```

```python
X = combined_clinical.iloc[:, 2:7]
y = combined_clinical.iloc[:, 1]
model = XGBClassifier()
model.fit(X, y)
plot_importance(model)
plt.show()
```

```python
xgboost_accuracy=[]
loo_accuracy=[]
i = 1
while i < 101:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=i)
    model = XGBClassifier()
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    xgboost_accuracy.append(accuracy)
    
    loo = LeaveOneOut()
    loo.get_n_splits(X)
    y_true = []
    y_est = []
    for train_index, test_index in loo.split(X):
        sub_train_index = np.random.choice(train_index, 20, replace=False)
        X_train, X_test = X.iloc[sub_train_index], X.iloc[test_index]
        y_train, y_test = y[sub_train_index], y[test_index]
        model = model.fit(X_train, y_train)
        y_est_1 = model.predict(X_test)
        y_true.append(y_test.iloc[0])
        y_est.append(y_est_1[0])
        
    accuracy = accuracy_score(y_true, y_est)
    loo_accuracy.append((accuracy.mean()*100.0))
    i += 1
    
    
print("Xgboost Accuracy: %.2f%%" % (sum(xgboost_accuracy) ))
print("Loocv Accuracy: %.2f%%" % (sum(loo_accuracy)/ 100.0))    
```
