```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np

from sklearn import metrics
from xgboost import XGBClassifier
from sklearn import svm

from numpy import sort
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import LeaveOneOut
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
import os
os.chdir('C://Users//YE/Desktop/')
```

```python
clinical_raman=pd.read_csv("clinical_raman.csv",header=None)
clinical_raman=clinical_raman.dropna(axis=1)
clinical_raman
```

```python
pca=PCA(0.95)
```

```python
x = StandardScaler().fit_transform(clinical_raman.iloc[:,2:])
y=pca.fit_transform(x) # principal component that could explain 95% of variability.
y=pd.DataFrame(y)
np.shape(y)
y
```

```python
pca.explained_variance_ratio_
```

```python
pd.concat([clinical_raman.iloc[:,:2],y],axis=1).to_csv("pca_clinical.csv")
```

```python
pca_clinical=pd.read_csv("pca_clinical.csv",header=None)
pca_clinical
```

```python
X = pca_clinical.iloc[:, 2:]
y = pca_clinical.iloc[:, 1]
xgboost_accuracy=[]
loo_accuracy=[]
i = 1
while i < 101:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=i)
    model = XGBClassifier()
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    xgboost_accuracy.append(accuracy)
    
    loo = LeaveOneOut()
    loo.get_n_splits(X)
    y_true = []
    y_est = []
    for train_index, test_index in loo.split(X):
        sub_train_index = np.random.choice(train_index, 20, replace=False)
        X_train, X_test = X.iloc[sub_train_index], X.iloc[test_index]
        y_train, y_test = y[sub_train_index], y[test_index]
        model = model.fit(X_train, y_train)
        y_est_1 = model.predict(X_test)
        y_true.append(y_test.iloc[0])
        y_est.append(y_est_1[0])
        
    accuracy = accuracy_score(y_true, y_est)
    loo_accuracy.append((accuracy.mean()*100.0))
    i += 1
    
    
print("Xgboost Accuracy: %.2f%%" % (sum(xgboost_accuracy) ))
print("Loocv Accuracy: %.2f%%" % (sum(loo_accuracy)/ 100.0))  
```

```python
warnings.filterwarnings('ignore')
X = pca_clinical.iloc[:, 2:]
y = pca_clinical.iloc[:, 1]
#X=np.array(X).reshape(len(X), -1)
X_scaled = preprocessing.scale(X)
```

```python
svm_accuracy=[]
loo_accuracy=[]

i=1

while i<101:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)
    clf = svm.SVC(kernel='linear', random_state=i)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    svm_accuracy.append(metrics.accuracy_score(y_test, y_pred))
    
  
    loo = LeaveOneOut()
    loo.get_n_splits(X)
    y_true = []
    y_est = []

    
    for train_index, test_index in loo.split(X):
        sub_train_index = np.random.choice(train_index, 20, replace=False)
        X_train, X_test = X.iloc[sub_train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[sub_train_index], y.iloc[test_index]

        model = clf.fit(X_train, y_train)
        y_est_1 = model.predict(X_test)
        y_true.append(y_test.iloc[0])
        y_est.append(y_est_1[0])

    accuracy = accuracy_score(y_true, y_est)

    loo_accuracy.append((accuracy.mean()*100.0))
    i += 1

print("SVM accuracy: "+str(sum(svm_accuracy)/100))
print("Loocv accuracy: "+str(sum(loo_accuracy)/10000))
```
