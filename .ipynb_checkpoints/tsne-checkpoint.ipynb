```python
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import umap
```

```python
import os
os.chdir('C://Users//YE/Desktop/')
```

```python
clinical=pd.read_excel("Clinicalversion.xlsx",index_col=0,header=1)
redclinical = clinical.loc[[936.12,1030.6,1031.7,1605.3,1660]]
```

```python
nonEoE_Tissue = redclinical.loc[:,"RS53":"RS42"]
ActiveEoe = redclinical.loc[:,"RS27":"BF14"]
InactiveEoE = redclinical.loc[:,"BF06":"RS52"]
EoE = pd.concat([nonEoE_Tissue,ActiveEoe,InactiveEoE],axis=1)
```

```python
x=TSNE(n_components=2,random_state=2).fit_transform(EoE.T)
plt.scatter(x[0:10,0],x[0:10,1],c="r",label="nonEoE_Tissue")
plt.scatter(x[10:18,0],x[10:18,1],c="g",label="ActiveEoE")
plt.scatter(x[18:24,0],x[18:24,1],c="b",label="InactiveEoE")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
x=TSNE(n_components=2,random_state=2).fit_transform(EoE.T)
plt.scatter(x[0:10,0],x[0:10,1],c="r",label="nonEoE_Tissue")
plt.scatter(x[10:18,0],x[10:18,1],c="g",label="ActiveEoE")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
reducer = umap.UMAP(random_state=2)
embedding = reducer.fit_transform(EoE.T)
plt.scatter(embedding[0:10,0],embedding[0:10,1],c="r",label="nonEoE_Tissue")
plt.scatter(embedding[10:18,0],embedding[10:18,1],c="g",label="ActiveEoE")
plt.scatter(embedding[18:24,0],embedding[18:24,1],c="b",label="InactiveEoE")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
